### 6. In class SLL, define a method insert_after() to insert a new_node after a given node of the list

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def insert_at_start(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_last(self, new_data):
        new_node = Node(new_data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def search(self, element):
        if self.is_empty():
            return -1  # List is empty
        temp = self.head
        counter = 0
        while temp:
            if temp.data == element:
                return counter
            temp = temp.next
            counter += 1
        return -1  # Element not found

    def insert_after(self, new_data, after_node):
        new_node=Node(new_data)
        if self.is_empty():
            self.head=new_node
        temp=self.head
        counter=0
        while temp:
            if counter==after_node:
                new_node.next=temp.next
                temp.next=new_node
                break
            temp=temp.next
            counter+=1


# Example usage
l = SinglyLinkedList()
l.insert_at_start(3)
l.insert_at_start(2)
l.insert_at_start(1)
l.insert_at_last(4)
print(l.search(4))  # Output: 3
print(l.search(5))  # Output: -1
l.insert_after(5,2)
```
